```python
import numpy as np
import matplotlib.pyplot as plt
```

# Less pythonic way (more loops)

```python
def mlpinit(w1):
    w2 = np.random.normal(0,1,10)
    b  = np.random.uniform(-2,2,10)
    
    return lambda x: w2.T @ np.tanh(w1*x - b)

def mlpplot(mlp):
    X = np.linspace(-2,2,70,endpoint=True)
    Y = [mlp(x) for x in X]
    
    plt.plot(X,Y)

plt.figure()
for i in range(50):
    mlp = mlpinit(np.random.normal(0,2,10))
    mlpplot(mlp)
plt.show()

plt.figure()
for i in range(50):
    mlp = mlpinit(np.random.normal(0,0.5,10))
    mlpplot(mlp)
plt.show()
```

# more pythonic way
a.,b.,c.)

```python
def mlp(w1):
    w2 = np.random.normal(0,1,10)
    b  = np.random.uniform(-2,2,10)
 
        
    y = lambda x: (w2 * np.tanh((x[:, np.newaxis] * w1 - b))).sum(axis=1)
    return y

x = np.linspace(-2,2,50,endpoint=True)

Y_b_uniform = [mlp(np.random.normal(0,2,10)) for _ in range(50)]
plt.figure()
for y in Y_b_uniform:
    plt.plot(x,y(x))
plt.show()


Y_b_normal = [mlp(np.random.normal(0,0.5)) for _ in range(50)]
plt.figure()
for y in Y_b_normal:
    plt.plot(x,y(x))
plt.show()
```

d.)

```python
g = lambda x: -x
mse = lambda a,b: (a-b)**2
```

```python
mse_uniform = np.array([mse(g(x),y(x)) for y in Y_b_uniform])
mse_uniform.sum(1).mean()
```

```python
mse_normal = np.array([mse(g(x),y(x)) for y in Y_b_normal])
mse_normal.sum(1).mean()
```

```python
y_uniform_best = Y_b_uniform[mse_uniform.sum(1).argmin()]
y_normal_best = Y_b_normal[mse_normal.sum(1).argmin()]

plt.plot(x, g(x), label='$g(x)$')
plt.plot(x, y_uniform_best(x), label='$b$ uniform')
plt.plot(x, y_normal_best(x), label='$b$ normal')
plt.legend()
```
