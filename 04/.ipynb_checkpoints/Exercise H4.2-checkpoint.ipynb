```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, -1], [1, 0.3], [1, 2]])
yt = np.array([-0.1, 0.5, 0.5])
w = np.array([-0.45, 0.2])

def learn(X, yt, w_old, step_function):
    y = X @ w_old

    error =  1/2 * (y - yt)**2
    error_prime = (y - yt)
#     print(error.mean())
    
    #                        delta  f(h)
    gradient_b = error_prime * (1)
    gradient_w = error_prime * (1 * X@w_old)

    gradient = np.vstack([gradient_b, gradient_w]).T
#     gradient = gradient.mean(0)
    
    print(step_function(X, gradient))
    w_new = w_old - step_function(X, gradient)
    return w_new

def step_gd(X, gradient):
    return 0.08 * gradient.mean(0)

def step_linesearch(X, gradient):
    result = calc_n(X, gradient) * gradient.mean(0) 
    return result.mean(0)

def calc_n(X, g):
    H = np.expand_dims(X, axis=2) @ np.expand_dims(X, axis=1)
    numerator = np.expand_dims(g, axis=1) @ np.expand_dims(g, axis=2)
    denominator = np.expand_dims(g, axis=1) @ H @ np.expand_dims(g, axis=2)
    
    return (numerator/denominator).mean(0) 

W = []
r = np.arange(10)
for i in r:
    w = learn(X, yt, w, step_linesearch)
    W.append(w)

W = np.array(W)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

ax1.scatter(W[:, 0], W[:, 1])
ax1.set_ylim(0.0, 0.5)
ax1.set_xlim(-0.5, 0.3)
ax1.set_title('Trace of weights using gradient descent')
ax1.set_xlabel('w_0')
ax1.set_ylabel('w_1')

ax2.plot(r, W[:, 0], label='w_0')
ax2.plot(r, W[:, 1], label='w_1')
ax2.set_title('Evolution of weights using gradient descent')
ax2.set_xlabel('iterations')
ax2.set_ylabel('weights')
ax2.legend()
```
